Извор за податоци ќе ми биде страната autoscout24.com која е страна за огласи за автомобили во Европа. Целта на овaa фаза е да извлечам податоци како што се производител, модел, година на производство, зафатнина на мотор, коњски сили, тип на гориво, тип на продавач и цена.  


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.autoscout24.com/lst?source=detailsearch'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```
